```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp_data = np.loadtxt(fname='T_Alicante_EM.csv', delimiter=',', skiprows=1)
```

Exercise 2.1: Plot the monthly and annual difference between max and min temperatures as a function of the month (1-12) and the year (1961-2096), respectively. In this case try to combine the plt.plot and plt.scatter functions. Hint: the plot function accept the syntax plt.plot(x,y).

```python
#finding the monthly difference
max_m = np.max(temp_data[:,1:], axis=0)
min_m = np.min(temp_data[:,1:], axis=0)
diff_m = max_m - min_m

months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
```

```python
figure1=plt.figure(figsize=(10,5))
plt.plot(months, diff_m)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Difference', fontsize=15)
plt.title('Monthly difference between maximum and minimum temperatures', fontsize=15)
```

```python
#finding the annual difference
max_a = np.max(temp_data[:, 1:], axis=1)
min_a = np.min(temp_data[:, 1:], axis=1)
diff_a = max_a - min_a
```

```python
figure2=plt.figure(figsize=(10,5))
plt.scatter(temp_data[:, 0:1], diff_a)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Difference', fontsize=15)
plt.title('Annual difference between maximum and minimum temperatures', fontsize=15)
```

Exercise 2.2: Plot the standard deviation of the monthly and annual temperatures as a function of the month (1-12) and the year (1961-2096), respectively. Hint: check the std function in NumPy.

```python
stdev_m = np.std(temp_data[:,1:], axis = 0) # monthly stdev
stdev_a = np.std(temp_data[:,1:], axis = 1) # annual stdev
```

```python
figure3=plt.figure(figsize=(10,5))
plt.plot(months, stdev_m)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Standard Deviation', fontsize=15)
plt.title('Standard deviation of monthly temperatures', fontsize=15)
```

```python
figure4=plt.figure(figsize=(10,5))
plt.scatter(temp_data[:, 0:1], stdev_a)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Standard Deviation', fontsize=15)
plt.title('Standard deviation of annual temperatures', fontsize=15)
```

Exercise 2.3: Prepare a plot with six panels (arranged as you wish) which depicts the annual dependence of the average monthly temperature for meteorological Spring (Mar, Apr, May) and Fall (Sep, Nov, Dec) seasons.

```python
fig, axs = plt.subplots(2, 3, figsize = (15, 10))
x = temp_data[:, 0:1]

# March
axs[0, 0].scatter(x, temp_data[:, 3])
axs[0, 0].set_title("March")

# April
axs[0, 1].scatter(x, temp_data[:, 4])
axs[0, 1].set_title("April")

# May
axs[0, 2].scatter(x, temp_data[:, 5])
axs[0, 2].set_title("May")

# September
axs[1, 0].scatter(x,temp_data[:, 9])
axs[1, 0].set_title("September")

# November
axs[1, 1].scatter(x, temp_data[:, 10])
axs[1, 1].set_title("November")

# December
axs[1, 2].scatter(x, temp_data[:, 11])
axs[1, 2].set_title("December")

plt.tight_layout()
plt.show()
```
